### *Seasonal forecast of fire danger indices from the Copernicus Emergency Management Service*

This notebook demonstrates how to access and visualise data from the CDS dataset "*Seasonal forecast of fire danger indices from the Copernicus Emergency Management Service*".

In this example, we will be using the `earthkit` Python package to access and visualise our data.

```python
import earthkit.data
import earthkit.maps
```

Data in the CDS can be accessed using the `earthkit.data.from_source` function. For our purposes, this function needs to take three arguments:
- The string `'cds'` - the *source* of our data
- The name of the dataset - in our case `'cems-fire-seasonal'`
- The payload of the request - the parameters we need to pass to access a specific part of the dataset.

Note that you will need a CDS account and a `.cdsapirc` file in order to have permission to access CDS data. If you have not set this up already, see the [CDS API user guide](https://cds.climate.copernicus.eu/api-how-to).

```python
data = earthkit.data.from_source(
    'cds', 'cems-fire-seasonal',
    {
        'format': 'netcdf',
        'release_version': '5',
        'variable': 'fire_weather_index',
        'year': '1991',
        'month': '09',
        'leadtime_hour': '12',
    },
)
```

We can use an xarray representation of our dataset to see that our request contains 25 ensemble members (the `number` dimension below).

```python
data.to_xarray()
```

Let's define a *style* with which we can visualise this data. The thresholds for Fire Weather Index categories are:
- 0 - very low
- 5.2 - low
- 11.2 - moderate
- 21.3 - high
- 38 - very high
- 50 - extreme

So let's create an **earthkit-maps** style with the above levels, and colours ranging from green to yellow to red to dark blue.

```python
style = earthkit.maps.Style(
    levels=[0, 5.2, 11.2, 21.3, 38, 50, 150],
    colors=['#84F07F', '#FFEB3C', '#FFB00C', '#FA4F00', '#B40000', '#280923'],
)
```

Finally, we can plot our data using the style defined above. For this demo, we are simply selecting ensemble member 0 (the control member) for our visualisation.

```python
earthkit.maps.quickplot(data[0], style=style)
```

Alternatively, we might want to define a style which labels the individual Fire Weather Index categories:

```python
disjoint_style = earthkit.maps.Style(
    levels=[0, 5.2, 11.2, 21.3, 38, 50, 150],
    colors=['#84F07F', '#FFEB3C', '#FFB00C', '#FA4F00', '#B40000', '#280923'],
    bin_labels=["very low", "low", "moderate", "high", "very high", "extreme"],
    legend_style="disjoint",
)

earthkit.maps.quickplot(data[0], style=disjoint_style)
```
